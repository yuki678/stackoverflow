```python
comp_country_group = df_result[['Country', 'ConvertedComp']].dropna().groupby(by='Country')
df_comp_country = comp_country_group.count().rename(columns={'ConvertedComp': 'Count'})
df_comp_mean = comp_country_group.mean().rename(columns={'ConvertedComp': 'Comp Mean (USD)'})
df_comp_country = df_comp_country.merge(df_comp_mean, on='Country')
df_comp_max = comp_country_group.max().rename(columns={'ConvertedComp': 'Comp Max (USD)'})
df_comp_country = df_comp_country.merge(df_comp_max, on='Country')
df_comp_min = comp_country_group.min().rename(columns={'ConvertedComp': 'Comp Min (USD)'})
df_comp_country = df_comp_country.merge(df_comp_min, on='Country')
df_comp_country.sort_values(by='Comp Mean (USD)', ascending=False)
```
